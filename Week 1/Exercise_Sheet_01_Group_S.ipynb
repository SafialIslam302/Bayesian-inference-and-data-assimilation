```python
p_0 =  -0.98909373  
q_0 =  -2.16120922
k = 1
point = int(20/0.2)

def euler_function(q, p, delta_t = 0.2):
    q_s = [q]
    p_s = p
    
    for _ in range(point - 1):
        q_n = q_s[-1] + (delta_t / k) * p_s
        p_s = p_s - (delta_t * k * q_s[-1])
        q_s.append(q_n)
    return q_s

euler = euler_function(p_0, q_0)
time_4 = np.arange(0, 20, 0.20)

plt.plot(time_4, euler, '-', color = "red", label = "estimate")
plt.plot(time, graph, '-', label = "truth")
plt.plot([1, 2], [-2, 0], 'o', label = "observed")
plt.xlabel("time")
plt.ylabel("position")

plt.legend()
```

Euler's technique is simple and quick to implement, but it requires small sizes for correct computations. The smaller the stepsize, the more steps, time, and computing cost are necessary to obtain a satisfying response [1].

Euler's technique involves reconstructing the function as a series of short, straight lines. It calculates the slope of these lines by assuming that the slope at a given location is the function's derivative at that point. As a result, it draws a line with the step size as the run and the slope as the rise. This small line's place to start is the end of the previous line segment. When one line segment comes to an end, Euler's method recalculates a new line at its new position until it reaches its final position. When the intervals are smaller, the slope is recalculated and updated at shorter time intervals. Euler's technique becomes increasingly accurate as the step sizes become smaller and the number of lines formed approaches infinity since a function is made up of an endless number of small lines [2].
