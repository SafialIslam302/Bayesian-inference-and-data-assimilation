# Problem 3 (ii)

```python
plt.clf()
X1 = []

dt = 0.01
N = 10000
fig1, ax1 = plt.subplots(1, 1, figsize=(12, 5))

for i in range(N+1):
    if i == 0:
        X1.append(4)
    
    if i > 0:
        X1.append(X1[i-1] - (dt * (((X1[i-1]-4)**2-2)*(4*X1[i-1]-16))) + np.sqrt(2*dt)* np.random.randn())
    
    if i in (0, 50, 100, 1000, 10000):
        if i == 0:
            ax1.hist(X1, bins = 30, color = "magenta", density = True, label="t = 0.0")
        if i == 50:
            ax1.hist(X1, bins = 30, color = "blue", density = True, label="t = 0.5")
        if i == 100:
            ax1.hist(X1, bins = 30, color = "red", density = True, label="t = 1.0")
        if i == 1000:
            ax1.hist(X1, bins = 30, color = "green", density = True, label="t = 10")
        if i == 10000:
            ax1.hist(X1, bins = 30, color = "orange", density = True, label="t = 100")
        plt.xlabel('x')
        plt.ylabel('p.d.f. of x')
        
        plt.ylim([0,3])
        
        plt.legend()
```

## Problem 3 (iii)

```python
import numpy

mus_001 = np.zeros(int(10/0.01))
sig_001 = np.zeros(int(10/0.01))

arr_X1 = numpy.array(X1)


for n in range(int(10/0.01)):
    mus_001[n] = np.mean(arr_X1[n])
    sig_001[n] = np.var(arr_X1[n])

import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot( np.arange(int(10/0.01))*0.01, mus_001, label = 'mean dt=0.01')
plt.plot( np.arange(int(10/0.01))*0.01, np.zeros(int(10/0.01)), 
ls = "--", label = 'mean = 0')
plt.xlabel('t')
plt.ylabel('Simulated Mean')
plt.legend(loc= "upper right")

plt.subplot(2,2,2)
plt.plot( np.arange(int(10/0.01))*0.01, sig_001, label = 'var dt=0.01')
plt.plot( np.arange(int(10/0.01))*0.01, np.zeros(int(10/0.01)), 
ls = "--",label = 'Var = 0.0904523')
plt.xlabel('t')
plt.ylabel('Simulated Var')
plt.legend(loc= "lower right")


# Some formatting
plt.gcf().set_size_inches(10,8)
plt.tight_layout()
plt.show()
```
