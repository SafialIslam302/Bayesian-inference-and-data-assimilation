```python
import numpy as np
from numpy.random import normal
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
a = np.array([1, 2, 3, 4])
w2 = np.array([5/8, 1/8, 1/8, 1/8])

N = 2000
mean_arr = np.zeros(N)
```

```python
def sample(M, L, w):
    eta = np.zeros(M)
    data = np.random.uniform(0, 1, L)

    for l in range(L):
        u = data[l]
        w_sum = w[0]

        i = 0
        while w_sum < u:
            i += 1
            w_sum += w[i]

        eta[i] += 1
        
    return eta
```

```python
for j in range(N):
    mean_arr[j] = np.mean(np.multiply(a, sample(4, 4, w2)))

print("Mean of the samples:")
print(np.mean(mean_arr))
print("Variance of the samples:")
print(np.var(mean_arr))
```
